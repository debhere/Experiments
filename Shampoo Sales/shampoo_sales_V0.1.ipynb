```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#plt.style.available
```

```python
plt.style.use('fivethirtyeight')
```

```python
shampoo = pd.read_csv("shampoo_sales.csv")
shampoo.sample(10)
```

```python
shampoo.info()
```

```python
shampoo[["Year","Month"]] = shampoo["Month"].str.split("-", expand=True)

shampoo["Year"] = shampoo["Year"].astype('category')
shampoo["Month"] = shampoo["Month"].astype('category')
```

```python
shampoo.sample(10)
```

```python
shampoo.info()
```

```python
shampoo["Month"].unique()
```

```python
shampoo["Year"].unique()
```

```python
sns.barplot(x="Year", y="Sales", data=shampoo)
sns.lineplot(x="Year", y="Sales", data=shampoo, errorbar=None)
plt.title("Yearly Sales Trend")
plt.show()
```

```python
sns.barplot(x="Month", y="Sales", data=shampoo)
plt.title("Monthly Shampoo Sales Distribution")
plt.show()
```

```python
ax = shampoo["Sales"].plot(title="Shampoo Sales in Last 3 Years", linewidth=2)
ax.axvline(x=11, color="red", linestyle="--", linewidth=0.7)
ax.axvline(x=23, color="red", linestyle="--", linewidth=0.7)
plt.show()
```
